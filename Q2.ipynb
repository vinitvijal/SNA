```python
import networkx as nx

G = nx.les_miserables_graph()
```

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
weights = [G[u][v]['weight'] for u, v in G.edges()]
nx.draw(G, pos, with_labels=True, edge_color=weights, edge_cmap=plt.cm.Oranges, node_color='lightgray')
plt.show()
```

To Normalise

```python
DG = G.to_directed()
```

```python
# For example, normalize weights
for u, v in DG.edges():
    DG[u][v]['weight'] = G[u][v]['weight'] / max(nx.get_edge_attributes(G, 'weight').values())
```
