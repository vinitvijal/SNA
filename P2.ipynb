### 2.	Compute and plot degree distribution of a real-world network. Also compute its local and global properties.

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Load real-world network (Les Misérables)
G = nx.les_miserables_graph()
```

```python
# Compute degree of each node
degrees = [G.degree(n) for n in G.nodes()]
unique_degrees = sorted(set(degrees))
degree_count = [degrees.count(d) for d in unique_degrees]
```

```python
# Plot degree distribution
plt.figure(figsize=(8, 6))
plt.bar(unique_degrees, degree_count, color='salmon')
plt.xlabel("Degree")
plt.ylabel("Number of Nodes")
plt.title("Degree Distribution of Les Misérables Network")
plt.grid(True)
plt.show()
```

```python
# Local Clustering Coefficient
clustering_coeffs = nx.clustering(G)

# Average Local Clustering Coefficient
avg_clustering = nx.average_clustering(G)

# Print a few examples
print("Local Clustering Coefficient (first 5 nodes):")
for node, coeff in list(clustering_coeffs.items())[:5]:
    print(f"{node}: {coeff:.3f}")
print(f"\nAverage Clustering Coefficient: {avg_clustering:.3f}")
```

```python
# Global Metrics
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
density = nx.density(G)
avg_degree = sum(degrees) / num_nodes
diameter = nx.diameter(G)
avg_shortest_path = nx.average_shortest_path_length(G)

print("\n--- Global Network Properties ---")
print(f"Number of Nodes: {num_nodes}")
print(f"Number of Edges: {num_edges}")
print(f"Network Density: {density:.4f}")
print(f"Average Degree: {avg_degree:.2f}")
print(f"Network Diameter: {diameter}")
print(f"Average Shortest Path Length: {avg_shortest_path:.3f}")
```
