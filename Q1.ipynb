```python
import networkx as nx
```

```python
G = nx.karate_club_graph()
```

```python
import matplotlib.pyplot as plt 
```

```python
DG = G.to_directed()
```

```python
DG.size()
```

```python
import random

for u, v in DG.edges():
    DG[u][v]['weight'] = random.uniform(0.5, 5)
```

```python
pos = nx.spring_layout(DG)
weights = [DG[u][v]['weight'] for u, v in DG.edges()]
nx.draw(DG, pos, with_labels=True, edge_color=weights, edge_cmap=plt.cm.Blues,)
plt.show()
```
